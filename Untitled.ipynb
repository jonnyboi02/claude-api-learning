```python
#load env variables

from dotenv import load_dotenv

load_dotenv()
```

```python
#create an api client

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python

def add_user_message (messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message (messages, text):
    assisant_message = {"role": "assistant", "content": text}
    messages.append(assisant_message)



def chat (messages):   
    message = client.messages.create(
        model = model,
        max_tokens = 1000,
        messages=messages
    )
    return message.content[0].text
```

```python
messages = []

startMessage = "Define Quantum computing in one sentence"

add_user_message(messages, startMessage)
add_assistant_message(messages, chat(messages))

followMessage = "Write another sentence"
add_user_message(messages, followMessage)
add_assistant_message(messages, chat(messages))

for i in messages:
    print(i)

```

```python
#exercise 1
def exercise1():
    messages = []
    userInput = input("Enter prompt")
    while userInput != "Stop":
        add_user_message(messages, userInput)
        assistantMessage = chat(messages)
        
        add_assistant_message(messages, assistantMessage)
        print(assistantMessage)
        userInput = input("Enter prompt")

exercise1()
```
